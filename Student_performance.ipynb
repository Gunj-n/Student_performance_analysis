```python
sns.countplot(x="gender",data=df)

plt.figure(figsize =(10, 3))
plt.show()
```

```python
This graph shows that female candidate are more than the male candidates.
```

```python
df['race/ethnicity'] = pd.Categorical(df['race/ethnicity'], categories=["group A", "group B", "group C", "group D", "group E"], ordered=True)
sns.countplot(x="race/ethnicity",hue="gender",data=df)
plt.show()
```

```python
This graph shows that on the basis of race/ethnicity group C category has more candidates than all other categories.And in that category females are more
than males.
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
color_dict = dict(zip(df['race/ethnicity'].unique(), colors))
pie_colors = [color_dict[race] for race in df['race/ethnicity'].value_counts().index]
plt.pie(x=df['race/ethnicity'].value_counts(), labels=df['race/ethnicity'].value_counts().index, explode=[0.1, 0, 0, 0, 0], autopct='%1.1f%%', shadow=True, colors=pie_colors, textprops={'fontsize': 14})
plt.show()
```

```python
This pie chart shows that group C has 31.9% of distribution which is the highest among all categories.
```

```python
plt.rcParams['figure.figsize'] = (12, 5)
sns.histplot(df["parental level of education"])
plt.title('Comparison of Parental Education', fontweight = 30, fontsize = 20)
plt.xlabel('Degree')
plt.ylabel('count')
plt.show()
```

```python
According to the comparison of parental education,the candidates are from those categories whose parents have completed their education till some college.
```

```python
fig, axes = plt.subplots(1, 3, figsize=(24, 6))
sns.violinplot(y='math score', data=df, linewidth=2, ax=axes[0])
axes[0].set_title('Distribution of Math Scores')
sns.violinplot(y='reading score', data=df, linewidth=2, ax=axes[1])
axes[1].set_title('Distribution of Reading Scores')
sns.violinplot(y='writing score', data=df, linewidth=2, ax=axes[2])
axes[2].set_title('Distribution of Writing Scores')
plt.tight_layout()
plt.show()
```

```python
These distributions shows that highest score in all the three subjects lies between 60-80 range.
```

```python
From the above dataset,we can conclude that female students performs better than male students.And their score in all the above mentioned subjects lies 
between 60-80 range.
```
